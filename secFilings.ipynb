```python
import requests
import time
import json

email = "rghielmini@macacobot.com"

firstYear = 1993
nextYear = 2023

forms = ['10-K', '10-Q', '20-F']
quarters = ['QTR1', 'QTR2', 'QTR3', 'QTR4']

years = [str(year) for year in range(firstYear,nextYear)]

sec = {
    form: dict() for form in forms
}

for year in years:
    for quarter in quarters:
        time.sleep(0.125) # 8 req/s: in order not to be rate limited (limit at 10 req/s)
        url = f"https://www.sec.gov/Archives/edgar/full-index/{year}/{quarter}/master.idx"
        response = requests.get(url, headers={'User-agent': email})
        if response.ok:
            print(f"{year}-{quarter}-OK")
            response.encoding = 'latin-1'
            afterHeaders = False
            for line in response.iter_lines():
                if line:
                    try:
                        line = line.decode()
                    except:
                        print(f"{year}-{quarter}-MISSING-LINE")
                        continue
                    entries = line.split("|")
                    if len(entries) == 5:
                        if afterHeaders:
                            cik = entries[0]
                            formType = entries[2]
                            filingDate = entries[3]
                            for _formType in sec.keys():
                                if formType == _formType:
                                    if filingDate not in sec[formType]:
                                        sec[formType][filingDate] = set()
                                    sec[formType][filingDate].add(cik)
                        else:
                            afterHeaders = True
        else:
            print(f"{year}-{quarter}-MISSING")

class SetEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, set):
            return list(obj)
        return json.JSONEncoder.default(self, obj)

json.dump(sec, open('secFilings.json', 'w'), cls=SetEncoder)
```
